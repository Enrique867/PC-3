```python
import pandas as pd

Data1 = "https://github.com/Enrique867/PC-3/raw/refs/heads/main/Education%20expenditures.csv"
Data2 = "https://github.com/Enrique867/PC-3/raw/refs/heads/main/Electricity%20-%20installed%20generating%20capacity.csv"
Data3 = "https://github.com/Enrique867/PC-3/raw/refs/heads/main/Energy%20consumption%20per%20capita.csv"
Data4 = "https://github.com/Enrique867/PC-3/raw/refs/heads/main/Reserves%20of%20foreign%20exchange%20and%20gold.csv"


df_education = pd.read_csv(Data1)
df_electricity = pd.read_csv(Data2)
df_energy = pd.read_csv(Data3)
df_reserves = pd.read_csv(Data4)
```

```python
df_education.head()
df_education.info()


```

```python
df_ordenado = df_education.sort_values(by='name', ascending=True)
# Convert the column to string type before using the .str accessor
df_ordenado['% of GDP'] = df_ordenado['% of GDP'].astype(str).str.replace(',', '')
df_ordenado['Education'] = pd.to_numeric(df_ordenado['% of GDP'], errors='coerce')
df_ordenado = df_ordenado.drop(columns=['% of GDP'])
df_ordenado.head()
```

```python
df_electricity.head()
df_electricity.info()
```

```python
df_ordenado1= df_electricity.sort_values(by='name', ascending=True)
df_ordenado1[' kW'] = df_ordenado1[' kW'].str.replace(',', '')
df_ordenado1['Electricity'] = pd.to_numeric(df_ordenado1[' kW'], errors='coerce')
df_ordenado1 = df_ordenado1.drop(columns=[' kW'])
df_ordenado1.head()
```

```python
df_energy.head()
df_energy.info()
```

```python
df_ordenado2= df_energy.sort_values(by='name', ascending=True)
df_ordenado2[' Btu/person'] = df_ordenado2[' Btu/person'].astype(str).str.replace(',', '')
df_ordenado2['Energy'] = pd.to_numeric(df_ordenado2[' Btu/person'], errors='coerce')
df_ordenado2 = df_ordenado2.drop(columns=[' Btu/person'])
df_ordenado2.head()
```

```python
df_reserves.head()
df_reserves.info()
```

```python
df_ordenado3= df_reserves.sort_values(by='name', ascending=True)
df_ordenado3['value'] = df_ordenado3['value'].astype(str).str.strip()
df_ordenado3['value'] = df_ordenado3['value'].str.replace('$', '',)
df_ordenado3['value'] = df_ordenado3['value'].str.replace(',', '')
df_ordenado3['value'] = df_ordenado3['value'].str.strip()
df_ordenado3['reserves'] = pd.to_numeric(df_ordenado3['value'], errors='coerce')
df_ordenado3.head()
```

```python
df_ordenado = df_ordenado[["name", "slug", "region", "Education"]]
df_ordenado1 = df_ordenado1[["name", "slug", "region", "Electricity"]]
df_ordenado2 = df_ordenado2[["name", "slug", "region", "Energy"]]
df_ordenado3 = df_ordenado3[["name", "slug", "region", "reserves"]]
```

```python
df_nuevo = pd.merge(df_ordenado, df_ordenado1, on=['name', 'region','slug'], how='inner')
df_nuevo = pd.merge(df_nuevo, df_ordenado2, on=['name', 'region','slug'], how='inner')
df_nuevo = pd.merge(df_nuevo, df_ordenado3, on=['name', 'region','slug'], how='inner')
df_nuevo
```

```python
df_resumen= df_nuevo.groupby('region')[['Education', 'Electricity', 'Energy', 'reserves']].agg(['mean','min','max'])
df_resumen
```

```python
from sklearn.preprocessing import RobustScaler
import mapclassify

vars_to_scale = ['reserves', 'Energy', 'Electricity']
scaler = RobustScaler(unit_variance=True, with_centering=False)
robScaled_data = scaler.fit_transform(df_nuevo[vars_to_scale])
vars_rob_names = [v + '_rob' for v in vars_to_scale]
df_nuevo[vars_rob_names] = robScaled_data
for var_rob in vars_rob_names:
    q5 = mapclassify.Quantiles(df_nuevo[var_rob], k=5)
    col_name = var_rob.replace('_rob', '') + '_level'
    df_nuevo[col_name] = q5.yb + 1
```

```python
k=5
df_nuevo['education_level'] = mapclassify.Quantiles(df_nuevo['Education'].dropna(), k=5).y.astype(int)+1
```
